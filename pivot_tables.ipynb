then calculate the min max cost of the tasks from the hourly cost of each task

CTO/Arquitectura	50,00 €

```python
cost = {
    "Desarrollo Frontend": 40,
    "Diseño UI/UX": 40,
    "CTO/Arquitectura": 50,
    "Desarrollo Backend": 50,
    "Prompting": 35,
    "Testeo": 35,
    "Reuniones": 0
}

# calculate the min max cost of the tasks from the hourly cost of each task
# calculate the min max cost of the tasks from the hourly cost of each task
df['Min Cost'] = df['Est. mín (h)'] * df['Tipo'].map(cost)
df['Max Cost'] = df['Est. máx (h)'] * df['Tipo'].map(cost)

df.head()
```

then create some pivot tables grouping first by Feature and then by Tipo

```python
# group by Feature and sum the min max hours and min max cost
Feature_df = df.groupby('Feature').agg({'Est. mín (h)': 'sum', 'Est. máx (h)': 'sum', 'Min Cost': 'sum', 'Max Cost': 'sum'})

# display the head
Feature_df
```

```python
# group by tipo and sum the min max hours and min max cost
tipo_df = df.groupby('Tipo').agg({'Est. mín (h)': 'sum', 'Est. máx (h)': 'sum', 'Min Cost': 'sum', 'Max Cost': 'sum'})

# display the head
tipo_df.head()
```

```python
# finally add up the final min max cost of all the project

print(f"The final min max cost of the project is {df['Min Cost'].sum()}€")
print(f"The final max cost of the project is {df['Max Cost'].sum()}€")
```
